```python
#function that takes 2 numbers as input
# and outputs their average
def avg(x,y):
    print ("first input is %d" % x)
    print ("second input is %d" % y)
    a = (x + y) / 2.0
    print ("average is %d" % a)
    return a
```

```python
avg(2,4)
```

```python
xvalues = linspace(-2.25, 0.75, 1000)
yvalues = linspace(-1.5, 1.5, 1000)

xlen = len(xvalues)
ylen = len(yvalues)

def mandel(c, maxiter):
    z = complex(0,0)
    for iteration in range(maxiter):
        z = (z*z) + c
        if abs(z) > 4:
            break
            pass
        pass
    return iteration

atlas = empty((xlen,ylen))
for ix in range(xlen):
    for iy in range(ylen):
        cx = xvalues[ix]
        cy = yvalues[iy]
        c = complex(cx,cy)
        atlas[ix,iy] = mandel(c,40)
        pass
    pass
figsize(18,18)

imshow(atlas.T, interpolation="nearest")
```
